```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import os
```

```python
df = pd.DataFrame()
for year in range(2015,2022):
    for month in range(1,13):
        filename=f"{year}-{month:02}_rws_filedata.csv"
        print(filename)
        
        newdf = pd.read_csv(f"datasets/TrafficJams/{filename}", sep=';')
        df = df.append(newdf)
```

```python
from datetime import datetime
monday = tuesday = wednesday = thursday = friday = 0
for date in df['DatumFileBegin'].unique():
    print(f"{date}: {len(df[df['DatumFileBegin'] == date].index)}")
        
    day = datetime.fromisoformat(date).weekday()
    if(day == 0):
        monday += len(df[df['DatumFileBegin'] == date].index)
        continue 
    elif(day == 1):
        tuesday += len(df[df['DatumFileBegin'] == date].index)
        continue 
    elif(day == 2):
        wednesday += len(df[df['DatumFileBegin'] == date].index)
        continue 
    elif(day == 3):
        thursday += len(df[df['DatumFileBegin'] == date].index)
        continue 
    elif(day == 4):
        friday += len(df[df['DatumFileBegin'] == date].index)
        continue 
    else:
        continue 
        
print(f"monday: {monday}")
print(f"tuesday: {tuesday}")
print(f"wednesday: {wednesday}")
print(f"thursday: {thursday}")
print(f"friday: {friday}")
```
