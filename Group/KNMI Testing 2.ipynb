```python
import pandas as pd
import numpy as np
import h5py
import math
import matplotlib.pyplot as plt
from pathlib import Path
import os
```

```python
path = 'C:/Users/ruben/OneDrive/Bureaublad/KNMI'
years = os.listdir(path)
months = os.listdir('{}/{}'.format(path, years[0]))
```

```python
files = []

for year in years:
    for month in months:
        for hour in os.listdir('{}/{}/{}'.format(path, year, month)):
            files.append(hour)
```

```python
def get_corresponding_hours(day):
    for i in range(1, 25):
        hour = day + str(i).rjust(2, '0')
        hours = [s for s in files if "RAD_NL25_RAC_MFBS_01H_" + day in s]
        print(hours)
```

```python
get_corresponding_hours("20080101")
```

```python
path = 'C:/Users/ruben/OneDrive/Bureaublad/KNMI/2008/01'
knmi = os.listdir(path)
ds = []
current_day = '01'
current_month = 'JAN'
for file in knmi:
    hdf = h5py.File('{}/{}'.format(path, file), 'r')
    starting_points = hdf['geographic'].attrs['geo_product_corners']
    lot_start = starting_points[3]
    lat_start = starting_points[2]
    date = hdf['overview'].attrs['product_datetime_end'][0].decode('UTF-8')
    year = date[7:11]
    month = date[3:6]
    day = date[0:2]
    hour = date[12:14]
    
    if(day != current_day):
        df = pd.DataFrame(ds, columns=['Precipitation', 'Longtitude', 'Latitude', 'Year', 'Month', 'Day', 'Hour'])
        df = df.drop(df[df.Precipitation <= 0].index)
        filepath = Path('C:/Users/ruben/OneDrive/Bureaublad/FHICT/Semester 4/Group Project/Datasets/KNMI/{}/{}/{}.csv'.format(year, current_month, current_day))  
        filepath.parent.mkdir(parents=True, exist_ok=True)
        df.to_csv(filepath,index=False, header=False)
        print(current_day)
        current_day = day
        ds = []
        
    data = np.array(hdf['image1/image_data'])
    for y in range(251, 580):
        for x in range(245, 501):
            value = data[y][x]
            ds.append([value, 
                       lot_start - y * 0.0086431372549019607843137254902, 
                       lat_start + x * 0.01550857142857142857142857142857,
                       year, month, day, hour])
print('Done')
```
