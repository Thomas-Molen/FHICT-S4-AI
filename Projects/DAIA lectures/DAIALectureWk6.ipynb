```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, month_plot, quarter_plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

```python
flights = sns.load_dataset("flights")
flights['report_date'] = flights.apply(lambda x: datetime.strptime(f"{x['year']}-{x['month']}", '%Y-%b').date(), axis=1)
flights.head()
```

```python
flights.plot(x='report_date', y='passengers')
# flights[flights['year'] == 1950].plot(x='report_date', y='passengers')
```

```python
sns.lineplot(data=flights, x='month', y='passengers', hue='year', legend='full')

plt.legend(reversed(plt.legend().legendHandles), reversed(flights['year']), loc='lower left', bbox_to_anchor=(1, 0))
#          ^ -->                                 ^ Reverse label order      ^ Set relative position   ^ Move lenged on x axis 
```
