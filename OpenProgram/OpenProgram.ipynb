##### What technical methods are used to create unsupervised recommendation systems
For this topic I found a very insightful [blog post](https://towardsdatascience.com/recommendation-systems-explained-a42fc60591ed).
I will be looking at the three systems explained in the blog post above, these being:
- Collaborative Filtering
- Content Based 
- Hybrid Recommendation

    ***Collaborative Filtering***  
    This method uses the principle that if people with similar behaviors are more likely to show interests in the same items in the future.  
    For Collaborative Filtering there are two approach: Memory and Model based.  
    
    Memory based is the more prominently discussed one, with this method similar products or people are grouped together and because of their similar behaviors will be expected to like the same things.  
    
    Model based takes the fundamental features from products and bases user's interests on the weight these features might have based on their previous experience.
    
    ***Content Based***  
    This methods uses a user's previously observed interests to suggest items with similar fundamental features. This means that if a user buys a lot of tech products it will be recommended more products from this category, or users who purchase high page count books will be recommended larger books in the future.
    
    ***Hybrid Recommendation***  
    As the name suggests this method allows for the combination of other techniques, most methods such as Collaborative and Content based have their disadvantages (for example content based won't work for new users). Hybrid Recommendation tries to get the best of both worlds by combining methods in either two ways:
    - Parallel
    - Sequential
    
        ***Parallel***  
        Parallel is the simpler of the two, it basically runs the different recommendation methods independently and combines the results.
        
        ***Sequential***  
        Sequential makes use of a large pool of possible items and feeds this list through each method one after the other, continuesly reducing the amount of items from the list untill the final recommendation output is created.

```python
refunds = df[df['Invoice'].astype('str').str.contains('C')]
print(f"Refund entries to remove: {len(refunds)}")
```

Note this takes a while

```python
import os
def RemoveRefunds(refund, df):
#   Find refund in original df
    rowIndex = df[(df['Customer ID'] == refunds.loc[refund.name]['Customer ID']) & 
     (df['StockCode'] == refunds.loc[refund.name]['StockCode']) & 
     (df['Price'] == refunds.loc[refund.name]['Price']) & 
     (df['Quantity'] == -refunds.loc[refund.name]['Quantity'])].head(1).index
    
#   Drop the found refund from original df
    df.drop(rowIndex, inplace=True)
    
#   Print to jupyter notebook
    os.write(1, bytes(f"({refund.name}) Removing Entry: {rowIndex}\n", encoding='utf-8'))
    
refunds.apply(axis=1, func=RemoveRefunds, df=df)
print('Done')
```

#### Model Data Preperation
Because most recommender models have specific data requirements, I will be having to convert the data stored into a structure that can be used fore modelling.  
The most important part for this is that most modelling libraries require a form of rating score to train the model on, this will allow for easier evaluation as the real rating can be compared to a predicted rating the model creates.  
The data used however, does not contain any form of ratings value, for this I will instead be using the quantity of products sold, this is a quantifiable feature to show how much a person likes the product.  
sale quantities however vary a lot within the data, because of this I will be excluding the largest outliers and Min/Max scaling the data so that it will conform to a good 1/10 rating scale.

```python
import pandas as pd
import numpy as np

cols=['Invoice', 'StockCode', 'Description', 'Quantity', 'InvoiceDate', 'Price', 'Customer ID', 'Country']
dtypes={
    'Invoice' : 'category',
    'StockCode' : 'category',
    'Description' : 'str',
    'Quantity' : 'float64',
    'Price' : 'float64',
    'Customer ID' : 'category',
    'Country' : 'category'
}
df = pd.read_csv("Datasets/invoices_prepared.csv", sep="|", parse_dates=['InvoiceDate'], usecols=cols, dtype=dtypes)
df.info()
```

Because of the way most recommender libraries work, they require a rating score from users to find products that customers want, my dataset however does not have this feature. However, I can create a version of a rating feature by using the quantity of purchased items by the user.

Create a dataframe that has a record for each instance of an item being bought by a unique user.

```python
rdf = pd.DataFrame(columns=['User', 'Item', 'Rating'])

for product in df['StockCode'].unique():
    rdf = rdf.append(pd.DataFrame({'User':df[df['StockCode'] == product]['Customer ID'].unique(), 'Item':product, 'Rating':0}))
```

Fill the rating column of each unique purchase with the quantity sold in total

```python
rdf['Rating'] = rdf.apply(axis=1, func=lambda x: df[(df['StockCode'] == x.Item) & (df['Customer ID'] == x.User)]['Quantity'].sum())
```

```python
rdf.to_csv("Datasets/model_data.csv", sep="|")
```

To make sure the scaling does not reduce the amount of perceived detail by squashing most numbers because of a couple outliers, I will be removing the outliers.

```python
srdf = rdf.copy()
srdf = srdf[srdf['Rating'] <= (rdf['Rating'].mean() + rdf['Rating'].std())]
```

Min/Max scaling the data will return values between a scale of 0-1, to give the model a better readability I will multiply the value by 10 to create a 0-10 scale.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
srdf[['Rating']] = pd.DataFrame(scaler.fit_transform(srdf[['Rating']]), columns=['Rating'])
srdf[['Rating']] = srdf[['Rating']]*10
```

```python
srdf.to_csv("Datasets/model_data_scaled.csv", sep="|")
```

Creating a recommendation system starts by using algorithms that will help the system, these algorithms can be collaborative filtering algorithms or other techniques such as Matrix Factorization.

I will be working towards implementing a collaborative filtering implementation, but first will try a more basic algorithm called NormalPredictor in scikit surprise, this will create a baseline to improve upon.
I will be doing all my modelling using cross_validation as to allow for good evaluation and get rid of possible data split biases.

```python
from surprise.reader import Reader
from surprise import NormalPredictor
from surprise import Dataset
from surprise.model_selection import cross_validate

# Create reader to parse pandas data
reader = Reader(line_format='user item rating', rating_scale=(0, srdf['Rating'].max()))
# Read pandas data into surprise dataformat
data = Dataset.load_from_df(srdf, reader)

# We'll use the famous SVD algorithm.
npr = NormalPredictor()

# Run 5-fold cross-validation and print results
cross_validate(npr, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

As seen from the results above, this model is already pretty good in predicting the Ratings score (which in reality is an identifier for the quantity the user will buy), it gets within an error of 1 with the prediction on a scale from 0-10, I would say that this is already fairly good, when it comes to finding the products that the customer is most likely to buy in large amounts.
